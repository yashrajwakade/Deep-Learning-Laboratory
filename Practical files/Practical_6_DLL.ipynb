# R-CNN Based Object Detection

## Aim
To implement Region-Based Convolutional Neural Network (R-CNN) for object detection and identify objects along with their locations in an image.


## Theory

### What is Object Detection?

Object Detection is a Computer Vision task that identifies:
1. What object is present?
2. Where the object is located?

The output contains:
- Object Class
- Bounding Box Coordinates
- Confidence Score

### What is R-CNN?

R-CNN stands for Region-Based Convolutional Neural Network.

It combines:
- Region Proposal Methods
- CNN Feature Extraction
- Classification

R-CNN was one of the first successful Deep Learning based object detection frameworks.

### R-CNN Architecture

Input Image
↓
Selective Search
↓
Region Proposals
↓
CNN Feature Extraction
↓
Classifier (SVM)
↓
Bounding Box Regression
↓
Final Detection

### Working Steps

1. Generate approximately 2000 region proposals.
2. Warp each region into a fixed size.
3. Pass regions through CNN.
4. Extract features.
5. Classify regions.
6. Refine bounding boxes.

### Advantages

- Good detection accuracy.
- Effective feature extraction.

### Limitations

- Slow training.
- Slow inference.
- Large storage requirements.


## Faster Variants of R-CNN

### Fast R-CNN
- Single CNN pass.
- Faster than R-CNN.

### Faster R-CNN
- Uses Region Proposal Network (RPN).
- Much faster.
- Higher accuracy.

Nowadays Faster R-CNN is widely used.


## Step 1: Import Libraries

```
import torch
import torchvision
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.transforms import functional as F

from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

## Step 2: Load Pretrained Faster R-CNN Model

```
model = fasterrcnn_resnet50_fpn(
    pretrained=True
)

model.eval()

print("Faster R-CNN Model Loaded Successfully")
```

## Step 3: Load Input Image

```
image = Image.open('sample.jpg')

plt.imshow(image)
plt.axis('off')
plt.show()
```

## Step 4: Convert Image to Tensor

```
img_tensor = F.to_tensor(image)

input_image = [img_tensor]
```

## Step 5: Perform Detection

```
with torch.no_grad():
    predictions = model(input_image)

print(predictions[0].keys())
```

## Step 6: COCO Class Labels

```
COCO_INSTANCE_CATEGORY_NAMES = [
'__background__','person','bicycle','car','motorcycle',
'airplane','bus','train','truck','boat','traffic light',
'fire hydrant','N/A','stop sign','parking meter','bench',
'bird','cat','dog','horse','sheep','cow','elephant',
'bear','zebra','giraffe','N/A','backpack','umbrella',
'N/A','N/A','handbag','tie','suitcase','frisbee','skis',
'snowboard','sports ball','kite','baseball bat',
'baseball glove','skateboard','surfboard','tennis racket',
'bottle','N/A','wine glass','cup','fork','knife','spoon',
'bowl','banana','apple','sandwich','orange','broccoli',
'carrot','hot dog','pizza','donut','cake','chair','couch',
'potted plant','bed','N/A','dining table','N/A','N/A',
'toilet','N/A','tv','laptop','mouse','remote','keyboard',
'cell phone','microwave','oven','toaster','sink',
'refrigerator','N/A','book','clock','vase','scissors',
'teddy bear','hair drier','toothbrush'
]
```

## Step 7: Visualize Detected Objects

```
fig, ax = plt.subplots(1, figsize=(12,8))
ax.imshow(image)

boxes = predictions[0]['boxes']
scores = predictions[0]['scores']
labels = predictions[0]['labels']

for box, score, label in zip(boxes, scores, labels):

    if score > 0.7:

        x1, y1, x2, y2 = box.numpy()

        rect = patches.Rectangle(
            (x1,y1),
            x2-x1,
            y2-y1,
            linewidth=2,
            fill=False
        )

        ax.add_patch(rect)

        ax.text(
            x1,
            y1,
            COCO_INSTANCE_CATEGORY_NAMES[label],
            fontsize=10
        )

plt.axis('off')
plt.show()
```

## Step 8: Print Detection Results

```
for score, label in zip(
    predictions[0]['scores'],
    predictions[0]['labels']
):

    if score > 0.7:
        print(
            "Object:",
            COCO_INSTANCE_CATEGORY_NAMES[label],
            " Confidence:",
            round(score.item(),3)
        )
```
