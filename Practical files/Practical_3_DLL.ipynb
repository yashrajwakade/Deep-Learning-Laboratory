# Multilayer Perceptron (MLP) for Classification

## Aim
To design and implement a Multilayer Perceptron (MLP) for classification and analyze its performance on a dataset.


## Theory

A Multilayer Perceptron (MLP) is a feedforward Artificial Neural Network consisting of:

1. Input Layer
2. One or More Hidden Layers
3. Output Layer

Unlike a Single Layer Perceptron, MLP can solve non-linear problems.

### Architecture

Input Layer → Hidden Layer(s) → Output Layer

### Components

#### Weights and Biases
Used to learn relationships from data.

#### Activation Functions

1. ReLU
f(x) = max(0,x)

2. Sigmoid
f(x)=1/(1+e^(-x))

3. Softmax
Used in multi-class classification.

### Training Process

1. Forward Propagation
2. Loss Calculation
3. Backpropagation
4. Weight Update

### Advantages

- Learns complex patterns
- Solves non-linear problems
- Widely used in classification tasks


## Dataset

The Iris Dataset is used.

Classes:
- Setosa
- Versicolor
- Virginica

Features:
- Sepal Length
- Sepal Width
- Petal Length
- Petal Width


## Step 1: Import Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, classification_report
```

## Step 2: Load Dataset

```
iris = load_iris()

X = iris.data
y = iris.target

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)

print("Classes:", iris.target_names)
```

## Step 3: Preprocessing

```
scaler = StandardScaler()
X = scaler.fit_transform(X)

y_cat = to_categorical(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_cat,
    test_size=0.2,
    random_state=42
)

print("Training Samples:", X_train.shape[0])
print("Testing Samples:", X_test.shape[0])
```

## Step 4: Build MLP Model

```
model = Sequential([
    Dense(16, activation='relu', input_shape=(4,)),
    Dense(8, activation='relu'),
    Dense(3, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

## Step 5: Train the Model

```
history = model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=8,
    validation_split=0.2,
    verbose=1
)
```

## Step 6: Evaluate Model

```
loss, accuracy = model.evaluate(
    X_test,
    y_test,
    verbose=0
)

print("Test Accuracy:", accuracy)
print("Test Loss:", loss)
```

## Step 7: Plot Accuracy Graph

```
plt.figure(figsize=(7,5))

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train','Validation'])

plt.show()
```

## Step 8: Plot Loss Graph

```
plt.figure(figsize=(7,5))

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train','Validation'])

plt.show()
```

## Step 9: Predictions

```
predictions = model.predict(X_test)

predicted_classes = np.argmax(predictions, axis=1)
actual_classes = np.argmax(y_test, axis=1)

print(classification_report(
    actual_classes,
    predicted_classes
))
```

## Step 10: Confusion Matrix

```
cm = confusion_matrix(
    actual_classes,
    predicted_classes
)

print("Confusion Matrix")
print(cm)
```

## Result Analysis

Observations:

1. MLP learns nonlinear decision boundaries.
2. Hidden layers extract meaningful features.
3. Accuracy improves with training epochs.
4. Backpropagation minimizes loss.
5. Softmax provides class probabilities.

Expected Accuracy:
95% - 100% on Iris Dataset.


## Result

The Multilayer Perceptron (MLP) model was successfully implemented using TensorFlow/Keras and trained on the Iris dataset. The model achieved high classification accuracy and correctly classified the majority of samples.
