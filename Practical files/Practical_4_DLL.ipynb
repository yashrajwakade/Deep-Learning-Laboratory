# Design and Implement a CNN Model to Classify an Image Dataset

## Aim
To design, implement, train, and evaluate a Convolutional Neural Network (CNN) for image classification using the CIFAR-10 dataset.


## Theory

### What is CNN?

Convolutional Neural Network (CNN) is a specialized Deep Learning architecture designed for image processing and computer vision tasks.

CNN automatically learns image features such as:
- Edges
- Corners
- Textures
- Shapes
- Objects

Unlike traditional Machine Learning, CNN performs automatic feature extraction.

### CNN Architecture

Input Image
↓
Convolution Layer
↓
ReLU Activation
↓
Pooling Layer
↓
Convolution Layer
↓
Pooling Layer
↓
Flatten
↓
Fully Connected Layer
↓
Output Layer

### Main Components

#### 1. Convolution Layer
Extracts important features using filters/kernels.

#### 2. ReLU Layer
Introduces non-linearity.

ReLU(x) = max(0,x)

#### 3. Pooling Layer
Reduces image dimensions and computation.

#### 4. Flatten Layer
Converts feature maps into a 1D vector.

#### 5. Dense Layer
Performs final classification.


## Dataset: CIFAR-10

The CIFAR-10 dataset contains:

- 60,000 images
- 10 classes
- Image size: 32×32×3

Classes:
1. Airplane
2. Automobile
3. Bird
4. Cat
5. Deer
6. Dog
7. Frog
8. Horse
9. Ship
10. Truck


## Step 1: Import Libraries

```
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

import matplotlib.pyplot as plt
import numpy as np

from sklearn.metrics import confusion_matrix, classification_report
```

## Step 2: Load Dataset

```
(X_train, y_train), (X_test, y_test) = datasets.cifar10.load_data()

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

## Step 3: Display Sample Images

```
class_names = [
    'airplane','automobile','bird','cat','deer',
    'dog','frog','horse','ship','truck'
]

plt.figure(figsize=(10,5))

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_train[i])
    plt.title(class_names[y_train[i][0]])
    plt.axis('off')

plt.show()
```

## Step 4: Normalize Images

```
X_train = X_train / 255.0
X_test = X_test / 255.0
```

## Step 5: Build CNN Model

```
model = models.Sequential()

model.add(layers.Conv2D(
    32,(3,3),
    activation='relu',
    input_shape=(32,32,3)
))

model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Conv2D(
    64,(3,3),
    activation='relu'
))

model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Conv2D(
    64,(3,3),
    activation='relu'
))

model.add(layers.Flatten())

model.add(layers.Dense(
    64,
    activation='relu'
))

model.add(layers.Dense(
    10,
    activation='softmax'
))

model.summary()
```

## Step 6: Compile Model

```
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

## Step 7: Train Model

```
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    validation_data=(X_test,y_test)
)
```

## Step 8: Evaluate Model

```
test_loss, test_acc = model.evaluate(
    X_test,
    y_test,
    verbose=0
)

print("Test Accuracy:", test_acc)
print("Test Loss:", test_loss)
```

## Step 9: Plot Accuracy Graph

```
plt.figure(figsize=(7,5))

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('CNN Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.legend(['Train','Validation'])

plt.show()
```

## Step 10: Plot Loss Graph

```
plt.figure(figsize=(7,5))

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('CNN Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.legend(['Train','Validation'])

plt.show()
```

## Step 11: Predictions

```
predictions = model.predict(X_test)

predicted_labels = np.argmax(
    predictions,
    axis=1
)

print("First Prediction:", predicted_labels[0])
print("Actual Label:", y_test[0][0])
```

## Step 12: Classification Report

```
print(
    classification_report(
        y_test,
        predicted_labels
    )
)
```

## Step 13: Confusion Matrix

```
cm = confusion_matrix(
    y_test,
    predicted_labels
)

print(cm)
```

## Result Analysis

Observations:

1. CNN automatically extracts image features.
2. Convolution layers detect edges and patterns.
3. Pooling reduces computational complexity.
4. Accuracy increases as training progresses.
5. CNN performs significantly better than traditional classifiers for images.

Expected Accuracy:
70% to 85% on CIFAR-10 using this basic architecture.


## Result

The Convolutional Neural Network (CNN) was successfully designed and implemented for image classification using the CIFAR-10 dataset. The model achieved good classification accuracy and successfully learned meaningful image features.
