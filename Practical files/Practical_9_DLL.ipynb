## Aim
To implement Video Classification using a CNN-RNN architecture and classify activities present in a video sequence.


## Theory

### What is Video Classification?

Video Classification is the process of assigning a category or label to an entire video.

Examples:
- Walking
- Running
- Jumping
- Dancing
- Playing Sports

### Why CNN-RNN?

A video contains:

#### Spatial Information
Information within a frame.

Handled by:
- CNN (Convolutional Neural Network)

#### Temporal Information
Information across frames over time.

Handled by:
- RNN
- LSTM
- GRU

### CNN-RNN Architecture

Video
↓
Frame Extraction
↓
CNN Feature Extraction
↓
Feature Sequence
↓
LSTM / RNN
↓
Dense Layer
↓
Video Class

### Working Principle

1. Extract frames from video.
2. Pass each frame through CNN.
3. Extract feature vectors.
4. Create feature sequence.
5. Feed sequence to LSTM.
6. Predict activity class.

### Advantages

- Learns spatial and temporal patterns.
- Effective for action recognition.
- Suitable for surveillance and activity analysis.

### Applications

- Human Activity Recognition
- Sports Analytics
- Surveillance Systems
- Autonomous Systems

## Step 2: Import Libraries

```
import cv2
import numpy as np
import tensorflow as tf

from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import TimeDistributed
from tensorflow.keras.layers import GlobalAveragePooling2D
```

## Step 3: Load Pretrained CNN

```
cnn_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

cnn_model.trainable = False

print("CNN Loaded Successfully")
```

## Step 4: Extract Frames from Video

```
video_path = 'sample_video.mp4'

cap = cv2.VideoCapture(video_path)

frames = []

while True:

    ret, frame = cap.read()

    if not ret:
        break

    frame = cv2.resize(frame,(224,224))

    frames.append(frame)

cap.release()

frames = np.array(frames)

print("Frames Shape:", frames.shape)
```

## Step 5: Normalize Frames

```
frames = frames / 255.0
```

## Step 6: Build CNN-RNN Model

```
model = Sequential([

    TimeDistributed(
        cnn_model,
        input_shape=(None,224,224,3)
    ),

    TimeDistributed(
        GlobalAveragePooling2D()
    ),

    LSTM(
        128,
        return_sequences=False
    ),

    Dense(
        64,
        activation='relu'
    ),

    Dense(
        5,
        activation='softmax'
    )
])

model.summary()
```

## Step 7: Compile Model

```
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

## Understanding CNN-RNN Pipeline

CNN Responsibilities:
- Edge Detection
- Texture Detection
- Feature Extraction

LSTM Responsibilities:
- Temporal Learning
- Sequence Modeling
- Motion Understanding

CNN learns "what appears in each frame".

LSTM learns "how frames change over time".


## CNN-RNN vs CNN

| Feature | CNN | CNN-RNN |
|----------|----------|----------|
| Image Classification | Yes | Yes |
| Video Classification | Limited | Excellent |
