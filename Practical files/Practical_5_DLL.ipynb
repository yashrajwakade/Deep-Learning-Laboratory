# Object Detection using CNN

## Aim
To implement Object Detection using a Convolutional Neural Network (CNN) based pre-trained model and identify objects present in an image.


## Theory

### What is Object Detection?

Object Detection is a Computer Vision task that not only identifies objects present in an image but also determines their locations using bounding boxes.

### Difference Between Classification and Detection

Image Classification:
- Predicts what object is present.

Object Detection:
- Predicts what object is present.
- Predicts where the object is located.

### CNN in Object Detection

CNN automatically extracts features such as:
- Edges
- Corners
- Textures
- Shapes

These features are used to detect objects in images.

### Object Detection Pipeline

Input Image
↓
Feature Extraction using CNN
↓
Region Proposal
↓
Object Classification
↓
Bounding Box Prediction
↓
Final Detection

### Applications

- Autonomous Vehicles
- Face Detection
- Medical Imaging
- Security Surveillance
- Traffic Monitoring


## Pre-trained Model Used

MobileNet SSD

Advantages:
- Fast
- Lightweight
- Good Accuracy
- Suitable for Real-Time Detection


## Step 1: Import Libraries

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2: Download Required Files

Download:

1. MobileNetSSD_deploy.prototxt
2. MobileNetSSD_deploy.caffemodel

Store them in the current working directory.


## Step 3: Define Class Labels

```
CLASSES = [
    "background", "aeroplane", "bicycle", "bird",
    "boat", "bottle", "bus", "car", "cat",
    "chair", "cow", "diningtable", "dog",
    "horse", "motorbike", "person",
    "pottedplant", "sheep", "sofa",
    "train", "tvmonitor"
]
```

## Step 4: Load Pre-trained CNN Model

```
net = cv2.dnn.readNetFromCaffe(
    "MobileNetSSD_deploy.prototxt",
    "MobileNetSSD_deploy.caffemodel"
)

print("Model Loaded Successfully")
```

## Step 5: Load Image

```
image = cv2.imread("sample.jpg")

image = cv2.resize(image,(600,400))

(h,w) = image.shape[:2]

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

## Step 6: Create Blob

```
blob = cv2.dnn.blobFromImage(
    cv2.resize(image,(300,300)),
    0.007843,
    (300,300),
    127.5
)

net.setInput(blob)
```

## Step 7: Perform Object Detection

```
detections = net.forward()

print("Detection Shape:", detections.shape)
```

## Step 8: Draw Bounding Boxes

```
for i in range(detections.shape[2]):

    confidence = detections[0,0,i,2]

    if confidence > 0.5:

        idx = int(detections[0,0,i,1])

        box = detections[0,0,i,3:7] * np.array(
            [w,h,w,h]
        )

        (startX,startY,endX,endY) = box.astype("int")

        label = f"{CLASSES[idx]} : {confidence:.2f}"

        cv2.rectangle(
            image,
            (startX,startY),
            (endX,endY),
            (0,255,0),
            2
        )

        cv2.putText(
            image,
            label,
            (startX,startY-10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.5,
            (0,255,0),
            2
        )
```

## Step 9: Display Detection Result

```
plt.figure(figsize=(10,6))

plt.imshow(
    cv2.cvtColor(
        image,
        cv2.COLOR_BGR2RGB
    )
)

plt.axis("off")
plt.show()
```
