## Aim
To implement Image Registration for aligning multimodal images and analyze the transformation between different image modalities.


## Theory

### What is Image Registration?

Image Registration is the process of aligning two or more images of the same scene captured:

- At different times
- From different viewpoints
- Using different sensors

The goal is to place corresponding pixels at the same spatial location.

### What is Multimodal Image Registration?

Multimodal registration aligns images obtained from different imaging modalities.

Examples:

- MRI and CT Scan
- PET and MRI
- RGB and Infrared Images
- Satellite Multispectral Images

### Why Image Registration?

Different sensors capture different information.

Registration combines complementary information from multiple modalities.

### Registration Pipeline

Reference Image
        +
Moving Image
        ↓
Feature Detection
        ↓
Feature Matching
        ↓
Transformation Estimation
        ↓
Image Warping
        ↓
Registered Image

### Applications

1. Medical Imaging
2. Remote Sensing
3. Robotics
4. Autonomous Vehicles
5. Surveillance Systems


## Types of Transformations

### Translation
Moves image horizontally or vertically.

### Rotation
Rotates image.

### Scaling
Changes image size.

### Affine Transformation
Combination of:
- Translation
- Rotation
- Scaling
- Shearing

### Perspective Transformation
Used for viewpoint changes.

## Step 2: Import Libraries

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Step 3: Load Images

```
reference_img = cv2.imread('reference.jpg')
moving_img = cv2.imread('moving.jpg')

reference_gray = cv2.cvtColor(
    reference_img,
    cv2.COLOR_BGR2GRAY
)

moving_gray = cv2.cvtColor(
    moving_img,
    cv2.COLOR_BGR2GRAY
)

print("Images Loaded Successfully")
```

## Step 4: Detect ORB Features

```
orb = cv2.ORB_create(5000)

kp1, des1 = orb.detectAndCompute(
    reference_gray,
    None
)

kp2, des2 = orb.detectAndCompute(
    moving_gray,
    None
)

print("Keypoints in Reference Image:", len(kp1))
print("Keypoints in Moving Image:", len(kp2))
```

## Step 5: Match Features

```
matcher = cv2.BFMatcher(
    cv2.NORM_HAMMING,
    crossCheck=True
)

matches = matcher.match(
    des1,
    des2
)

matches = sorted(
    matches,
    key=lambda x:x.distance
)

print("Total Matches:", len(matches))
```

## Step 6: Visualize Feature Matching

```
matched_img = cv2.drawMatches(
    reference_img,
    kp1,
    moving_img,
    kp2,
    matches[:50],
    None
)

plt.figure(figsize=(15,8))
plt.imshow(
    cv2.cvtColor(
        matched_img,
        cv2.COLOR_BGR2RGB
    )
)
plt.axis('off')
plt.show()
```

## Step 7: Estimate Transformation Matrix

```
src_pts = np.float32([
    kp1[m.queryIdx].pt
    for m in matches[:100]
]).reshape(-1,1,2)

dst_pts = np.float32([
    kp2[m.trainIdx].pt
    for m in matches[:100]
]).reshape(-1,1,2)

H, mask = cv2.findHomography(
    dst_pts,
    src_pts,
    cv2.RANSAC,
    5.0
)

print("Homography Matrix:")
print(H)
```

## Step 8: Perform Image Registration

```
height, width = reference_img.shape[:2]

registered_img = cv2.warpPerspective(
    moving_img,
    H,
    (width,height)
)

print("Image Registration Completed")
```

## Step 9: Display Registered Image

```
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(
    cv2.cvtColor(
        reference_img,
        cv2.COLOR_BGR2RGB
    )
)
plt.title("Reference Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(
    cv2.cvtColor(
        registered_img,
        cv2.COLOR_BGR2RGB
    )
)
plt.title("Registered Image")
plt.axis('off')

plt.show()
```
