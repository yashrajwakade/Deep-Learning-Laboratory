## Aim
To implement YOLO (You Only Look Once) for real-time object detection and identify objects along with their locations and confidence scores.


## Theory

### What is YOLO?

YOLO stands for **You Only Look Once**. It is a state-of-the-art object detection algorithm that detects multiple objects in a single pass through a neural network.

Unlike R-CNN based methods, YOLO treats object detection as a single regression problem.

### Why YOLO?

Traditional approaches:
- R-CNN
- Fast R-CNN
- Faster R-CNN

These perform:
1. Region Proposal
2. Feature Extraction
3. Classification

YOLO performs all operations in a single forward pass, making it extremely fast.

### YOLO Architecture

Input Image
↓
CNN Backbone
↓
Feature Maps
↓
Bounding Box Prediction
↓
Class Prediction
↓
Confidence Score
↓
Non-Max Suppression
↓
Final Detection

### Key Concepts

#### Bounding Box
Represents object location.

#### Confidence Score
Represents certainty of prediction.

#### Intersection over Union (IoU)
Measures overlap between predicted and actual boxes.

#### Non-Max Suppression (NMS)
Removes duplicate detections.

### Advantages

- Real-time detection
- High speed
- Good accuracy
- Detects multiple objects simultaneously

### Applications

- Autonomous Vehicles
- Surveillance Systems
- Traffic Monitoring
- Robotics
- Medical Imaging

## Step 1: Install Required Libraries

## Step 2: Import Libraries

```
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

## Step 3: Load YOLOv8 Pretrained Model

```
model = YOLO('yolov8n.pt')

print("YOLOv8 Model Loaded Successfully")
```

## Step 4: Perform Object Detection

```
results = model('sample.jpg')

print("Detection Completed")
```

## Step 5: Display Detection Results

```
for result in results:
    print(result.boxes)
```

## Step 6: Visualize Detected Objects

```
annotated_image = results[0].plot()

plt.figure(figsize=(12,8))
plt.imshow(cv2.cvtColor(
    annotated_image,
    cv2.COLOR_BGR2RGB
))
plt.axis('off')
plt.show()
```

## Step 7: Print Detected Objects

```
for result in results:

    boxes = result.boxes

    for box in boxes:

        class_id = int(box.cls[0])
        confidence = float(box.conf[0])

        print(
            "Class ID:",
            class_id,
            " Confidence:",
            round(confidence,3)
        )
```

## Understanding the Output

Output contains:

1. Bounding Box Coordinates
2. Object Class
3. Confidence Score

Example:

Person : 0.95

Car : 0.91

Dog : 0.88
