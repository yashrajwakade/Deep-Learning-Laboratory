# Video Classification using Vision Transformer (ViT)

## Aim
To implement Video Classification using a Vision Transformer (ViT) based architecture and classify activities present in a video sequence.


## Theory

### What is Video Classification?

Video Classification is the process of assigning a category label to an entire video based on its visual content.

Examples:
- Walking
- Running
- Playing Football
- Dancing
- Swimming

### Why Transformers for Videos?

Traditional CNN-RNN models:
- CNN extracts spatial features.
- RNN/LSTM learns temporal information.

Limitations:
- Difficulty learning long-range dependencies.
- Sequential processing is slow.

Vision Transformers solve these issues using Self-Attention.

### Video Vision Transformer (ViViT)

ViViT stands for Video Vision Transformer.

It extends Vision Transformers from images to videos.

### ViViT Architecture

Video
↓
Frame Extraction
↓
Patch Generation
↓
Patch Embedding
↓
Positional Encoding
↓
Transformer Encoder
↓
Classification Head
↓
Video Class

### Important Concepts

#### Patch Embedding
Each frame is divided into image patches.

#### Positional Encoding
Provides spatial and temporal information.

#### Self-Attention
Learns relationships among all patches across all frames.

#### Transformer Encoder
Processes complete video information simultaneously.

### Advantages

- Captures long-range dependencies.
- Better temporal understanding.
- Highly scalable.
- State-of-the-art performance.

### Applications

- Human Activity Recognition
- Sports Analytics
- Surveillance
- Medical Video Analysis
- Autonomous Systems

## Step 2: Import Libraries

```
import cv2
import torch
import numpy as np

from transformers import VivitImageProcessor
from transformers import VivitForVideoClassification
```

## Step 3: Load Pretrained ViViT Model

```
processor = VivitImageProcessor.from_pretrained(
    "google/vivit-b-16x2-kinetics400"
)

model = VivitForVideoClassification.from_pretrained(
    "google/vivit-b-16x2-kinetics400"
)

print("ViViT Model Loaded Successfully")
```

## Step 4: Extract Frames from Video

```
video_path = "sample_video.mp4"

cap = cv2.VideoCapture(video_path)

frames = []

while True:

    ret, frame = cap.read()

    if not ret:
        break

    frame = cv2.cvtColor(
        frame,
        cv2.COLOR_BGR2RGB
    )

    frames.append(frame)

cap.release()

print("Total Frames:", len(frames))
```

## Step 5: Sample Frames for Prediction

```
sampled_frames = frames[:32]

print("Frames Used:", len(sampled_frames))
```

## Step 6: Preprocess Video Frames

```
inputs = processor(
    sampled_frames,
    return_tensors="pt"
)

print(inputs.keys())
```

## Step 7: Perform Video Classification

```
with torch.no_grad():
    outputs = model(**inputs)

logits = outputs.logits

predicted_class_idx = logits.argmax(-1).item()

print("Predicted Class Index:", predicted_class_idx)
```

## Step 8: Display Predicted Label

```
predicted_label = model.config.id2label[
    predicted_class_idx
]

print("Predicted Activity:", predicted_label)
```

## Step 9: Display Top Predictions

```
probabilities = torch.nn.functional.softmax(
    logits,
    dim=-1
)

top5 = torch.topk(
    probabilities,
    5
)

for score, idx in zip(
    top5.values[0],
    top5.indices[0]
):

    print(
        model.config.id2label[idx.item()],
        "->",
        round(score.item(),4)
    )
```

## Understanding ViViT Pipeline

Unlike CNN-RNN:

## CNN-RNN vs ViViT

| Parallel Processing | No | Yes |
| Long-Range Dependencies | Limited | Excellent |
| Training Complexity | Lower | Higher |
| Accuracy | Good | Better |
