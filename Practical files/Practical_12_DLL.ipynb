# Stable Diffusion with Hugging Face

## Aim
To implement Stable Diffusion using Hugging Face Diffusers and generate images from textual prompts.


## Theory

### What is Stable Diffusion?

Stable Diffusion is a Text-to-Image Generative AI model that creates realistic images from natural language prompts.

Example:

Prompt:
"A futuristic city at night"

Output:
AI-generated image matching the description.

### What is a Diffusion Model?

Diffusion models learn by:

1. Adding noise to images.
2. Learning how to remove the noise.
3. Generating new images from random noise.

### Stable Diffusion Architecture

Text Prompt
↓
Tokenizer
↓
CLIP Text Encoder
↓
Latent Space
↓
UNet Denoising Network
↓
VAE Decoder
↓
Generated Image

### Main Components

#### CLIP
Converts text into semantic embeddings.

#### Latent Space
Compressed representation of images.

#### UNet
Gradually removes noise from latent representations.

#### VAE (Variational Autoencoder)
Converts latent representation into final image.

### Why Stable Diffusion?

- High image quality
- Fast generation
- Open source
- Customizable

## Step 2: Import Libraries

```
import torch

from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt
```

## Step 3: Load Stable Diffusion Model

```
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")

print("Stable Diffusion Model Loaded Successfully")
```

## Step 4: Define Prompt

```
prompt = "A futuristic smart city with flying cars"

print("Prompt:", prompt)
```

## Step 5: Generate Image

```
image = pipe(
    prompt,
    num_inference_steps=50,
    guidance_scale=7.5
).images[0]

print("Image Generated Successfully")
```

## Step 6: Display Generated Image

```
plt.figure(figsize=(8,8))

plt.imshow(image)
plt.axis('off')

plt.show()
```

## Step 7: Save Generated Image

```
image.save(
    "generated_image.png"
)

print("Image Saved Successfully")
```

## Understanding the Generation Process

Step 1:
Convert text prompt into embeddings using CLIP.

Step 2:
Generate random noise.

Step 3:
UNet gradually removes noise.

Step 4:
Generate latent image representation.

Step 5:
VAE converts latent image into final image.

Step 6:
Output generated image.


## Important Parameters

### num_inference_steps

Controls image quality.

Example:

20 Steps → Faster

50 Steps → Better Quality

### guidance_scale

Controls prompt adherence.

Lower:
More creative

Higher:
More prompt-focused

Typical value:
7.5
