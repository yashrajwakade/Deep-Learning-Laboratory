## Aim

## Theory

### What is Vision Transformer (ViT)?

Vision Transformer (ViT) is a Deep Learning architecture that applies the Transformer model, originally developed for Natural Language Processing (NLP), to image classification tasks.

Unlike CNNs, ViT processes an image as a sequence of image patches.

### Why ViT?

CNN:
- Learns local features.
- Uses convolution operations.

ViT:
- Learns global relationships.
- Uses self-attention mechanism.

### ViT Architecture

Input Image
↓
Image Patching
↓
Patch Embedding
↓
Positional Encoding
↓
Transformer Encoder
↓
MLP Head
↓
Classification Output

### Important Components

#### Image Patches
The image is divided into small patches.

Example:

224×224 Image
↓
16×16 Patches
↓
196 Tokens

#### Patch Embedding
Converts image patches into vectors.

#### Positional Encoding
Adds spatial information.

#### Self-Attention
Learns relationships among all image patches.

#### MLP Head
Produces final classification output.

### Advantages

- Captures global context.
- State-of-the-art performance.
- Scales effectively with large datasets.

### Applications

- Image Classification
- Medical Imaging
- Satellite Image Analysis
- Autonomous Vehicles

## Step 2: Import Libraries

```
from transformers import ViTImageProcessor
from transformers import ViTForImageClassification

from PIL import Image
import torch
import matplotlib.pyplot as plt
```

## Step 3: Load Pretrained ViT Model

```
processor = ViTImageProcessor.from_pretrained(
    'google/vit-base-patch16-224'
)

model = ViTForImageClassification.from_pretrained(
    'google/vit-base-patch16-224'
)

print("Vision Transformer Loaded Successfully")
```

## Step 4: Load Input Image

```
image = Image.open('sample.jpg')

plt.imshow(image)
plt.axis('off')
plt.show()
```

## Step 5: Preprocess Image

```
inputs = processor(
    images=image,
    return_tensors='pt'
)

print(inputs.keys())
```

## Step 6: Perform Image Classification

```
with torch.no_grad():
    outputs = model(**inputs)

logits = outputs.logits

predicted_class_id = logits.argmax(-1).item()

print("Predicted Class ID:", predicted_class_id)
```

## Step 7: Display Predicted Class

```
predicted_label = model.config.id2label[
    predicted_class_id
]

print("Predicted Label:", predicted_label)
```

## Step 8: Display Top Predictions

```
probabilities = torch.nn.functional.softmax(
    logits,
    dim=-1
)

top5 = torch.topk(
    probabilities,
    5
)

for score, idx in zip(
    top5.values[0],
    top5.indices[0]
):
    print(
        model.config.id2label[idx.item()],
        "->",
        round(score.item(),4)
    )
```

## Understanding ViT Working

1. Image is divided into patches.
2. Patches become tokens.
3. Positional information is added.
4. Transformer encoder learns patch relationships.
5. Classification head predicts image class.

Unlike CNNs, ViT can understand long-range dependencies among image regions.


## CNN vs ViT

| Feature | CNN | ViT |
|----------|----------|----------|
| Feature Learning | Local | Global |
| Convolution | Yes | No |
| Attention | No | Yes |
| Scalability | Medium | High |
| Training Data Requirement | Lower | Higher |
