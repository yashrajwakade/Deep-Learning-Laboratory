```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("traffic.jpg")

gray = cv2.cvtColor(
    image,
    cv2.COLOR_BGR2GRAY
)
```

```python
plt.figure(figsize=(6,6))

plt.imshow(
    gray,
    cmap='gray'
)

plt.title(
    "Original Image"
)

plt.axis("off")

plt.show()
```

```python
edge_kernel = np.array([

    [-1,-1,-1],
    [-1, 8,-1],
    [-1,-1,-1]

])
```

```python
blur_kernel = np.array([

    [1,1,1],
    [1,1,1],
    [1,1,1]

])/9
```

```python
edge_image = cv2.filter2D(
    gray,
    -1,
    edge_kernel
)
```

```python
sharpen_kernel = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

sharp_image = cv2.filter2D(
    gray,
    -1,
    sharpen_kernel
)
```

```python
blur_image = cv2.filter2D(
    gray,
    -1,
    blur_kernel
)
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)

plt.imshow(
    gray,
    cmap='gray'
)

plt.title("Original")

plt.axis("off")

# Edge

plt.subplot(2,2,2)

plt.imshow(
    edge_image,
    cmap='gray'
)

plt.title("Edge Detection")

plt.axis("off")

# Sharpen

plt.subplot(2,2,3)

plt.imshow(
    sharp_image,
    cmap='gray'
)

plt.title("Sharpened")

plt.axis("off")

# Blur

plt.subplot(2,2,4)

plt.imshow(
    blur_image,
    cmap='gray'
)

plt.title("Blurred")

plt.axis("off")

plt.show()
```
